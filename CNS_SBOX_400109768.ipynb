# Question1

```python
def change_one_bit(input,position):
  mask = 1 << position
  return mask ^ input

def sbox_output(input,sbox_index):
  input = bin(input)[2:].zfill(6)
  column = input[1:5]
  row = input[0]+input[5]
  #print("input = ",input)
  #print("row = ",int(row, 2))
  #print("column = ",int(column, 2))
  output = sbox[sbox_index][int(row, 2)][int(column, 2)]
  #print("output = ",output)
  return output

def dist(input1,input2):
  diff = input1 ^ input2
  diff = bin(diff)[2:].zfill(6)
  return diff.count('1') , diff



def change_one_bit_dist(input,position,sbox_index):
  input1 = input
  input2 = change_one_bit(input,position)
  #print("input1 = ",bin(input1)[2:].zfill(6))
  #print("input2 = ",bin(input2)[2:].zfill(6))
  output1 = sbox_output(input1,sbox_index)
  output2 = sbox_output(input2,sbox_index)
  #print("output1 = ",bin(output1)[2:].zfill(6))
  #print("output2 = ",bin(output2)[2:].zfill(6))

  return dist(output1,output2)
```

```python
def change_one_bit_table(sbox_index):
  field_names  = [""]
  rows = np.empty((6, 65), dtype='U10')
  distances = np.zeros((6, 64))
  for j in range(6):
    rows[j][0] = str("bit "+str(j+1))
  for i in range(64):
    input = i
    field_names = np.append(field_names,str(bin(input)[2:].zfill(6)))
    for q in range(6):
      distances[q][i],rows[q][i+1] = change_one_bit_dist(input,q,sbox_index)


  print("SBox (",sbox_index+1,") :")
  print("Minimum distance in Table :",np.min(distances))

  table = PrettyTable()
  table.field_names = list(dict.fromkeys(field_names).keys())
  for l in range(6):
    table.add_row(rows[l])
  print(table)
  print()
```

```python
for i in range(8):
  change_one_bit_table(i)
```

```python
def xor_input_table(sbox_index):
  field_names  = ["INPUTS"]
  input_xors = np.empty(65, dtype='U11')
  output_input = np.empty(65, dtype='U11')
  output_xor_input = np.empty(65, dtype='U11')
  diff = np.empty(65, dtype='U11')
  input_xors[0] = "X+001100"
  output_input[0] = "S(X)"
  output_xor_input[0] = "S(X+001100)"
  diff[0] = "Difference"
  distances = np.zeros(64)

  mask = int("001100", 2)

  for i in range(64):
    input = i
    field_names = np.append(field_names,str(bin(input)[2:].zfill(6)))
    input_xors[i+1] = input ^ mask
    output_input[i+1] = sbox_output(input,sbox_index)
    output_xor_input[i+1] = sbox_output(int(input_xors[i+1]),sbox_index)
    distances[i],diff[i+1] = dist(int(output_input[i+1]),int(output_xor_input[i+1]))


    input_xors[i+1] = str(bin(int(input_xors[i+1]))[2:].zfill(6))
    output_input[i+1] = str(bin(int(output_input[i+1]))[2:].zfill(6))
    output_xor_input[i+1] = str(bin(int(output_xor_input[i+1]))[2:].zfill(6))

  print("SBox (",sbox_index+1,") :")
  print("Minimum value in Table :",np.min(distances))

  distt = ["Distance"]
  distt = np.append(distt,np.asarray(distances, dtype=int)[:])


  table = PrettyTable()
  table.field_names = list(dict.fromkeys(field_names).keys())
  table.add_row(input_xors)
  table.add_row(output_input)
  table.add_row(output_xor_input)
  table.add_row(diff)
  table.add_row(distt)
  print(table)
  print()
```

```python
for i in range(8):
  xor_input_table(i)
```
